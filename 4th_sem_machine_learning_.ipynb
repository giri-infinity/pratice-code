```python
import numpy as np

epsilon = 0.0000001

def step_func(t):
    if t > epsilon:
        return 1
    else:
        return 0

X = np.array([
    [160, 56, 1],
    [163, 43, 1],
    [165, 48, 1],
    [170, 80, 1],
    [175,76,1],
    [180,70,1]
])

y = np.array([0, 0,0,1,1,1])
W = np.zeros(len(X[1]))

def perceptron_fit(X, Y, epochs=10):
    global W
    eta = 0.2
    for t in range(epochs):
        print(f"epoch{t+1}")
        for i in range(len(X)):
            predict = step_func(np.dot(X[i], W))
            error = Y[i] - predict
            W = W + eta * error * X[i]
            print(f"input:{X[i][0]},target:{y[i]},prediction:{predict},updated weights:{W}")
        print("================================")

def perceptron_predict(X, Y):
    global W
    for x in X:
        result = step_func(np.dot(x, W))
        gender = "male" if result == 1 else "female"
        print(f"{x[0]:.2f},{x[1]:.2f}->predicted gender:{gender}")

        new_data_points = [
        [172, 65, 1],
        [158, 50, 1],
        [185, 90, 1],
        [168, 60, 1]
    ]
    for data_point in new_data_points:
        result = step_func(np.dot(data_point, W))
        gender = "male" if result == 1 else "female"
        print(f"{data_point[0]:.2f},{data_point[1]:.2f}->predicted gender:{gender}")


perceptron_fit(X, y, 40)
perceptron_predict(X, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
x = np.arange(-10, 10, 0.1)
y = sigmoid(x)

plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.title("Sigmoid Function")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    if x > 0:
        return x
    else:
        return 0

x = np.arange(-10, 10, 0.1)
y = np.array([relu(val) for val in x])

plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.title("ReLU Function")
plt.grid(True)
plt.show()
```
