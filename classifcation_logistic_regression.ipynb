```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
x=np.arange(-5,5,0.1)
y=sigmoid(x)
plt.plot(x,y)
plt.plot()
print(y)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x + 5))
x=np.arange(-5,5,0.1)
y=sigmoid(x)
plt.plot(x,y)
plt.plot()
print(y)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
scores = np.array([65, 73, 85, 90, 98])
pass_or_fail = np.array([0, 0, 0, 1, 1])
model = LogisticRegression()
model.fit(scores.reshape(-1, 1), pass_or_fail)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-0.71813664 * x +62.8349207))
x=np.arange(60,100,10)
y=sigmoid(x)
plt.plot(x,y)
plt.plot()
```

```python
test=[[87],[88],[92]]
model.predict(test)
```

```python
test=[[87],[88],[92]]
model.predict_proba(test)
```
