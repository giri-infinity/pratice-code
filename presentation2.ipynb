```python
import numpy as np

def step_func(t):
    return 1 if t > 0 else 0

def perceptron_fit(X, y, epochs=50):
    W = np.zeros(X.shape[1])  # Initialize weights
    eta = 0.2  # Learning rate
    for epoch in range(epochs):
        print("Epoch =", epoch, "======================")
        for i in range(len(X)):
            prediction = step_func(np.dot(X[i], W))
            error = y[i] - prediction
            W += eta * error * X[i]  # Update weights
            print("Input =", X[i], "Label =", y[i], "Predicted =", prediction, "Updated Weights =", W)
        print("================================")
    return W

def perceptron_predict(X, W):
    return [step_func(np.dot(x, W)) for x in X]

# Training dataset (Runs, Batting Avg, Strike Rate, Wickets, Bias term=1)
X = np.array([
    [1753, 29.71, 141.8, 0, 1],
    [1522, 29.84, 126.3, 0, 1],
    [948, 18.96, 160.98, 38, 1],
    [716, 24.68, 161.54, 24, 1],
    [453, 10.53, 154.6, 79, 1],
    [0, 0.00, 0.0, 30, 1],
    [850, 18.50, 120.0, 50, 1],
    [120, 6.00, 70.0, 45, 1],
    [70, 4.50, 85.0, 121, 1],
    [620, 15.50, 110.0, 5, 1],
    [1000, 25.00, 112.0, 10, 1],
    [1100, 30.00, 90.0, 15, 1],
    [90, 9.00, 100.0, 12, 1],
    [300, 12.50, 98.0, 22, 1],
    [540, 18.00, 92.0, 5, 1],
    [310, 10.30, 101.0, 3, 1],
    [480, 16.50, 95.0, 2, 1],
    [800, 20.00, 89.0, 8, 1],
    [250, 12.00, 88.0, 6, 1],
    [190, 11.20, 85.0, 7, 1]
])

y = np.array([0, 0, 1, 1, 1, 0, 1, 0, 1, 0,
              0, 0, 0, 1, 0, 1, 0, 1, 0, 0])

# Train perceptron
W = perceptron_fit(X, y, epochs=500)

# Predict on training data
predictions = perceptron_predict(X, W)
print("Training Predictions:", predictions)

# New data for prediction (Runs, Batting Avg, Strike Rate, Wickets, Bias term=1)
new_X = np.array([
    [1200, 28.0, 130.0, 0, 1],   # Batsman
    [1500, 31.0, 140.0, 0, 1],   # Batsman
    [700, 15.0, 110.0, 40, 1],   # Bowler
    [400, 10.0, 90.0, 70, 1],    # Bowler
    [600, 12.0, 85.0, 80, 1]     # Bowler
])

new_predictions = perceptron_predict(new_X, W)
print("New Data Predictions:", new_predictions)
```
